### **Đề 1:** HỆ THỐNG QUẢN LÝ NHÂN VIÊN SỬ DỤNG STACK VÀ KẾ THỪA
- Các lớp Nhân viên:

    - Lớp Nhân viên (Employee): 
        - Lưu trữ thông tin cơ bản (id, tên, tuổi, email, lương cơ bản)
        - Cung cấp hàm tạo và phương thức biểu diễn chuỗi phù hợp
    - Lớp Lập trình viên (Developer) (kế thừa từ Employee): 
        - Thêm thuộc tính: ngôn ngữ lập trình chính, số năm kinh nghiệm, cấp độ 
        (junior/senior)
        - Ghi đè phương thức biểu diễn chuỗi
    - Lớp Quản lý Dự án (ProjectManager) (kế thừa từ Employee): 
        - Thêm thuộc tính: số dự án đã hoàn thành, lĩnh vực chuyên môn, chứng chỉ
        - Ghi đè phương thức biểu diễn chuỗi
    - Lớp Kiểm thử viên (Tester) (kế thừa từ Employee): 
        - Thêm thuộc tính: loại kiểm thử (manual/automation), công cụ sử dụng
        - Ghi đè phương thức biểu diễn chuỗi

- Lớp Stack:
    - Triển khai cấu trúc dữ liệu Stack với các phương thức sau: 
        - push: Thêm một mục vào stack
        - pop: Xóa và trả về mục trên cùng
        - is_empty: Kiểm tra xem stack có trống không
        - show: Hiển thị nội dung stack
- Lớp Quản lý Nhân viên (EmployeeManagement):
    - Sử dụng lớp Stack để quản lý nhân viên
- Triển khai các chức năng sau:

- Chức năng yêu cầu:
    - Thêm các loại nhân viên khác nhau vào hệ thống
    - Xóa nhân viên
    - Hiển thị danh sách nhân viên
    - Tìm kiếm nhân viên theo ID
    - Cập nhật thông tin nhân viên
    - Tìm nhân viên có kinh nghiệm cao nhất
    - Tìm quản lý dự án có nhiều dự án thành công nhất
    - Sắp xếp nhân viên theo mức lương sử dụng Bubble Sort
    - Triển khai chức năng hoàn tác (undo) các thao tác quản lý nhân viên bằng Stack
    - Lưu dữ liệu nhân viên vào file
    - Đọc dữ liệu nhân viên từ file
---

```python
from abc import ABC, abstractmethod

class Stack:
    def __init__(self):
        self.stack = []
        
    def is_empty(self):
        return len(self.stack) == 0
    
    def show(self):
        if self.is_empty():
            return "Stack is empty"
        else:
            return str(self.stack)
    
    def push(self, item):
        self.stack.insert(0, item)
    
    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop(0)

def display_menu():
    print("""
        1. Thêm các loại nhân viên khác nhau vào hệ thống
        2. Xóa nhân viên
        3. Hiển thị danh sách nhân viên
        4. Tìm kiếm nhân viên theo ID
        5. Cập nhật thông tin nhân viên
        6. Tìm nhân viên có kinh nghiệm cao nhất
        7. Tìm quản lý dự án có nhiều dự án thành công nhất
        8. Sắp xếp nhân viên theo mức lương sử dụng Bubble Sort
        9. Triển khai chức năng hoàn tác (undo) các thao tác quản lý nhân viên bằng Stack
        10. Lưu dữ liệu nhân viên vào file
        11. Đọc dữ liệu nhân viên từ file
        12. Thoát chương trình
        """)

class Employee(ABC):
    
    @abstractmethod
    def __init__(self, name, age=None, email=None, salary=None):
        self.name = name
        self.age = age
        self.email = email
        self.salary = salary

    def inThongTin(self):
        return f"Name: {self.name}, Age: {self.age}, Email: {self.email}, Lương cơ bản: {self.salary}"

class Developer(Employee):
    def __init__(self, name, age, email, salary, ngonNguLapTrinhChinh, namKinhNghiem, chungChi):
        super().__init__(name, age, email, salary)
        self.ngonNguLapTrinhChinh = ngonNguLapTrinhChinh
        self.namKinhNghiem = namKinhNghiem
        self.chungChi = chungChi
    
    def inThongTin(self):
        return super().inThongTin() + f", Ngôn ngữ lập trình: {self.ngonNguLapTrinhChinh}, Kinh nghiệm: {self.namKinhNghiem}, Chứng chỉ: {self.chungChi}"

class ProjectManager(Employee):
    def __init__(self, name, age, email, salary, soDuAnDaHoanThanh, linhVucChuyenMon, chungChi):
        super().__init__(name, age, email, salary)
        self.soDuAnDaHoanThanh = soDuAnDaHoanThanh
        self.linhVucChuyenMon = linhVucChuyenMon
        self.chungChi = chungChi

    def inThongTin(self):
        return super().inThongTin() + f", Số dự án đã hoàn thành: {self.soDuAnDaHoanThanh}, Lĩnh vực chuyên môn: {self.linhVucChuyenMon}, Chứng chỉ: {self.chungChi}"

class Tester(Employee):
    def __init__(self, name, age, email, salary, soBugDaTim, phamViKiemThao):
        super().__init__(name, age, email, salary)
        self.soBugDaTim = soBugDaTim
        self.phamViKiemThao = phamViKiemThao

    def inThongTin(self):
        return super().inThongTin() + f", Số bug đã tìm: {self.soBugDaTim}, Phạm vi kiểm thử: {self.phamViKiemThao}"

class EmployeeManagement:
    def __init__(self):
        self.listEmployee = []
        self.filename = "employee.txt"
        self.undo_stack = Stack()

    def add_employee(self, employee):
        self.listEmployee.append(employee)
        self.undo_stack.push(employee)  # Push the action onto the stack

    def remove_employee(self, employee):
        if employee in self.listEmployee:
            self.listEmployee.remove(employee)
            self.undo_stack.push(employee)  # Push the action onto the stack

    def display_employees(self):
        if not self.listEmployee:
            print("Danh sách nhân viên trống.")
            return
        for emp in self.listEmployee:
            print(emp.inThongTin())

    def find_employee_by_id(self, email):
        for emp in self.listEmployee:
            if emp.email == email:
                print(emp.inThongTin())
                return
        print("Không tìm thấy nhân viên với ID:", email)

    def update_employee_info(self, email, new_info):
        for emp in self.listEmployee:
            if emp.email == email:
                emp.name = new_info.get('name', emp.name)
                emp.age = new_info.get('age', emp.age)
                emp.salary = new_info.get('salary', emp.salary)
                print("Thông tin nhân viên đã được cập nhật.")
                return
        print("Không tìm thấy nhân viên với ID:", email)

    def find_most_experienced_employee(self):
        if not self.listEmployee:
            print("Danh sách nhân viên trống.")
            return
        most_experienced = max(self.listEmployee, key=lambda emp: getattr(emp, 'namKinhNghiem', 0))
        print("Nhân viên có kinh nghiệm cao nhất:", most_experienced.inThongTin())

    def find_top_project_manager(self):
        if not self.listEmployee:
            print("Danh sách nhân viên trống.")
            return
        top_manager = max((emp for emp in self.listEmployee if isinstance(emp, ProjectManager)),
                          key=lambda emp: emp.soDuAnDaHoanThanh, default=None)
        if top_manager:
            print("Quản lý dự án có nhiều dự án thành công nhất:", top_manager.inThongTin())
        else:
            print("Không có quản lý dự án nào trong danh sách.")

    def bubble_sort_employees_by_salary(self):
        n = len(self.listEmployee)
        for i in range(n):
            for j in range(0, n-i-1):
                if self.listEmployee[j].salary > self.listEmployee[j+1].salary:
                    self.listEmployee[j], self.listEmployee[j+1] = self.listEmployee[j+1], self.listEmployee[j]
        print("Đã sắp xếp nhân viên theo mức lương.")

    def undo_last_action(self):
        last_action = self.undo_stack.pop()
        if last_action:
            self.listEmployee.remove(last_action)
            print("Đã hoàn tác hành động cuối cùng.")
        else:
            print("Không có hành động nào để hoàn tác.")

    def save_file(self):
        if not self.listEmployee:
            print("Danh sách trống, không có gì để lưu.")
            return False

        try:
            with open(self.filename, 'w', encoding='utf-8') as file:
                for emp in self.listEmployee:
                    if isinstance(emp, Developer):
                        file.write(f"D|{emp.name} {emp.age} {emp.email} {emp.salary} {emp.ngonNguLapTrinhChinh} {emp.namKinhNghiem} {emp.chungChi}\n")
                    elif isinstance(emp, ProjectManager):
                        file.write(f"P|{emp.name} {emp.age} {emp.email} {emp.salary} {emp.soDuAnDaHoanThanh} {emp.linhVucChuyenMon} {emp.chungChi}\n")
                    elif isinstance(emp, Tester):
                        file.write(f"T|{emp.name} {emp.age} {emp.email} {emp.salary} {emp.soBugDaTim} {emp.phamViKiemThao}\n")
            print("Đã lưu dữ liệu vào file:", self.filename)
            return True
        except Exception as e:
            print("Lỗi: " + str(e))
            return False

    def load_file(self):
        try:
            with open(self.filename, 'r', encoding='utf-8') as file:
                for line in file:
                    data = line.strip().split()
                    if data[0] == 'D':
                        emp = Developer(data[1], int(data[2]), data[3], float(data[4]), data[5], int(data[6]), data[7])
                    elif data[0] == 'P':
                        emp = ProjectManager(data[1], int(data[2]), data[3], float(data[4]), int(data[5]), data[6], data[7])
                    elif data[0] == 'T':
                        emp = Tester(data[1], int(data[2]), data[3], float(data[4]), int(data[5]), data[6])
                    self.listEmployee.append(emp)
            print("Đã tải dữ liệu từ file:", self.filename)
            return True
        except Exception as e:
            print("Lỗi: " + str(e))
            return False
```

```python
from abc import ABC, abstractmethod

# Định nghĩa lớp Node cho LinkedList
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head is None

    def push(self, item):
        new_node = Node(item)
        new_node.next = self.head
        self.head = new_node

    def pop(self):
        if self.is_empty():
            return None
        popped_item = self.head.data
        self.head = self.head.next
        return popped_item

    def show(self):
        current = self.head
        items = []
        while current:
            items.append(current.data.inThongTin())  # Gọi phương thức inThongTin để hiển thị
            current = current.next
        return items if items else "LinkedList is empty"

    def search(self, email):
        current = self.head
        while current:
            if current.data.email == email:
                return current.data  # Trả về nhân viên nếu tìm thấy
            current = current.next
        return None  # Trả về None nếu không tìm thấy

    def delete(self, email):
        current = self.head
        previous = None
        while current:
            if current.data.email == email:
                if previous:
                    previous.next = current.next
                else:
                    self.head = current.next  # Nếu xóa phần tử đầu tiên
                return True  # Trả về True nếu xóa thành công
            previous = current
            current = current.next
        return False  # Trả về False nếu không tìm thấy


class EmployeeManagement:
    def __init__(self):
        self.listEmployee = []
        self.filename = "employee.txt"
        self.undo_list = LinkedList()  # Sử dụng LinkedList thay vì Stack

    def add_employee(self, employee):
        self.listEmployee.append(employee)
        self.undo_list.push(employee)  # Push hành động lên LinkedList

    def remove_employee_by_email(self, email):
        # Tìm và xóa nhân viên từ danh sách chính
        for emp in self.listEmployee:
            if emp.email == email:
                self.listEmployee.remove(emp)
                if self.undo_list.delete(email):  # Xóa từ LinkedList
                    print(f"Đã xóa nhân viên với email: {email}")
                return
        print(f"Không tìm thấy nhân viên với email: {email}")

    def display_employees(self):
        if not self.listEmployee:
            print("Danh sách nhân viên trống.")
            return
        for emp in self.listEmployee:
            print(emp.inThongTin())

    def find_employee_by_email(self, email):
        employee = self.undo_list.search(email)  # Sử dụng phương thức search của LinkedList
        if employee:
            print("Nhân viên tìm thấy:", employee.inThongTin())
        else:
            print("Không tìm thấy nhân viên với email:", email)

    def update_employee_info(self, email, new_info):
        for emp in self.listEmployee:
            if emp.email == email:
                emp.name = new_info.get('name', emp.name)
                emp.age = new_info.get('age', emp.age)
                emp.salary = new_info.get('salary', emp.salary)
                print("Thông tin nhân viên đã được cập nhật.")
                return
        print("Không tìm thấy nhân viên với ID:", email)

    def undo_last_action(self):
        last_action = self.undo_list.pop()
        if last_action:
            self.listEmployee.remove(last_action)
            print("Đã hoàn tác hành động cuối cùng.")
        else:
            print("Không có hành động nào để hoàn tác.")

    def save_file(self):
        if not self.listEmployee:
            print("Danh sách trống, không có gì để lưu.")
            return False

        try:
            with open(self.filename, 'w', encoding='utf-8') as file:
                for emp in self.listEmployee:
                    if isinstance(emp, Developer):
                        file.write(f"D|{emp.name} {emp.age} {emp.email} {emp.salary} {emp.ngonNguLapTrinhChinh} {emp.namKinhNghiem} {emp.chungChi}\n")
                    elif isinstance(emp, ProjectManager):
                        file.write(f"P|{emp.name} {emp.age} {emp.email} {emp.salary} {emp.soDuAnDaHoanThanh} {emp.linhVucChuyenMon} {emp.chungChi}\n")
                    elif isinstance(emp, Tester):
                        file.write(f"T|{emp.name} {emp.age} {emp.email} {emp.salary} {emp.soBugDaTim} {emp.phamViKiemThao}\n")
            print("Đã lưu dữ liệu vào file:", self.filename)
            return True
        except Exception as e:
            print("Lỗi: " + str(e))
            return False

    def load_file(self):
        try:
            with open(self.filename, 'r', encoding='utf-8') as file:
                for line in file:
                    data = line.strip().split()
                    if data[0] == 'D':
                        emp = Developer(data[1], int(data[2]), data[3], float(data[4]), data[5], int(data[6]), data[7])
                    elif data[0] == 'P':
                        emp = ProjectManager(data[1], int(data[2]), data[3], float(data[4]), int(data[5]), data[6], data[7])
                    elif data[0] == 'T':
                        emp = Tester(data[1], int(data[2]), data[3], float(data[4]), int(data[5]), data[6])
                    self.listEmployee.append(emp)
            print("Đã tải dữ liệu từ file:", self.filename)
            return True
        except Exception as e:
            print("Lỗi: " + str(e))
            return False
        
def main():  
    management = EmployeeManagement()  

    while True:  
        print("\n--- Hệ thống quản lý nhân viên ---")  
        print("1. Thêm nhân viên")  
        print("2. Xóa nhân viên")  
        print("3. Tìm kiếm nhân viên")  
        print("4. Cập nhật thông tin nhân viên")  
        print("5. Hiển thị danh sách nhân viên")  
        print("6. Hoàn tác hành động cuối cùng")  
        print("7. Lưu dữ liệu vào file")  
        print("8. Tải dữ liệu từ file")  
        print("0. Thoát")  

        choice = input("Xin nhập số để chọn thao tác: ")  

        if choice == '1':  
            # Thêm nhân viên  
            name = input("Nhập tên nhân viên: ")  
            age = int(input("Nhập tuổi: "))  
            email = input("Nhập email: ")  
            salary = float(input("Nhập lương: "))  
            role = input("Nhập vai trò (Developer/ProjectManager/Tester): ")  

            if role.lower() == 'developer':  
                language = input("Nhập ngôn ngữ lập trình chính: ")  
                experience = int(input("Nhập năm kinh nghiệm: "))  
                certificate = input("Nhập chứng chỉ: ")  
                emp = Developer(name, age, email, salary, language, experience, certificate)  
            elif role.lower() == 'projectmanager':  
                projects_completed = int(input("Nhập số dự án đã hoàn thành: "))  
                specialization = input("Nhập lĩnh vực chuyên môn: ")  
                certificate = input("Nhập chứng chỉ: ")  
                emp = ProjectManager(name, age, email, salary, projects_completed, specialization, certificate)  
            elif role.lower() == 'tester':  
                bugs_found = int(input("Nhập số bug đã tìm: "))  
                testing_scope = input("Nhập phạm vi kiểm thử: ")  
                emp = Tester(name, age, email, salary, bugs_found, testing_scope)  
            else:  
                print("Vai trò không hợp lệ.")  
                continue  

            management.add_employee(emp)  
            print("Đã thêm nhân viên.")  

        elif choice == '2':  
            email = input("Nhập email nhân viên cần xóa: ")  
            management.remove_employee_by_email(email)  

        elif choice == '3':  
            email = input("Nhập email nhân viên cần tìm: ")  
            management.find_employee_by_email(email)  

        elif choice == '4':  
            email = input("Nhập email nhân viên cần cập nhật: ")  
            new_info = {}  
            new_info['name'] = input("Nhập tên mới (nhấn Enter để giữ nguyên): ") or None  
            new_info['age'] = input("Nhập tuổi mới (nhấn Enter để giữ nguyên): ")  
            if new_info['age']:  
                new_info['age'] = int(new_info['age'])  
            new_info['salary'] = input("Nhập lương mới (nhấn Enter để giữ nguyên): ")  
            if new_info['salary']:  
                new_info['salary'] = float(new_info['salary'])  
            management.update_employee_info(email, {k: v for k, v in new_info.items() if v is not None})  

        elif choice == '5':  
            management.display_employees()  

        elif choice == '6':  
            management.undo_last_action()  

        elif choice == '7':  
            management.save_file()  

        elif choice == '8':  
            management.load_file()  

        elif choice == '0':  
            print("Thoát khỏi hệ thống quản lý.")  
            break  

        else:  
            print("Lựa chọn không hợp lệ. Xin vui lòng thử lại.")  

if __name__ == "__main__":  
    main()  
```
