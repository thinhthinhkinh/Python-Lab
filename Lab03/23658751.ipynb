### **Quản lí nhân viên**
- Thuộc tính:
    - ID (string)
    - Name (string)
    - Age (int)
    - Salary (double)

- Xây dựng chương trình bao gồm các chức năng sau:
    1. Đọc từ file employee.txt
    2. Thêm nhân viên vào danh sách
    3. Hiển thị danh sách nhân viên
    4. Tìm kiếm thông tin nhân viên theo mã (ID)
        - Tìm kiếm tuyến tính (Linear Search): Duyệt tuần tự để tìm nhân viên theo ID
        - Tìm kiếm nhị phân (Binary Search): Tìm kiếm nhanh trên danh sách đã sắp xếp theo 
    5. Xóa nhân viên khỏi danh sách 
    6. Cập nhật thông tin nhân viên
    7. Tính tuổi và hiển thị nhân viên có tuổi lớn nhất
    8. Sắp xếp nhân viên tăng dần theo lương (Bubble sort)
    9. Sắp xếp nhân viên theo năm sinh (Insertion sort)
    10. Sắp xếp danh sách nhân viên theo ID (Selection sort)
    10. Lưu thông tin vào file employee_new.txt
    11. Thoát

- Định dạng file:
    - File input (employee.txt): mỗi dòng chứa thông tin 1 nhân viên
    - Format: ID,Name,BirthYear,Salary
    - Ví dụ: E001,Nguyen Van A,1990,10000000
